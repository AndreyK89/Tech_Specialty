#### Очистка данных.

```python
# Нулевые качественные значения NaN в столбцах заменим на строку 'none'.
df[['Position', 'Nationality', 'Club', 'Contract Valid Until']] = df[['Position', 'Nationality', 'Club', 'Contract Valid Until']]. fillna('none')

# Количественные значения равные 0 (241), заполним mean значением.
df['Wage'].replace(0,df['Wage'].mean(),inplace=True)

# Нулевые количественные значения NaN(257) в столбце Value заменить на значение mean.
df['Value'] = df['Value']. fillna(df['Value'].mean())
df[['Release Clause']] = df[['Release Clause']]. fillna (35)

# Нулевые количественные значения NaN в столбцах заменим на значение min 1. 
df[['Skill Moves', 'International Reputation',]] = df[['Skill Moves', 'International Reputation']]. fillna (1)
```

### Задача №2 Найдите в датафрейме полные дубликаты и удалите их.  
Значения могут быть одинаковыми, но написаны по-разному.  
Например, может отличаться размер регистра (заглавные и строчные буквы). Особое внимание уделить колонке с названиями команд.

```python
# Переводим строковые значения столбцов в нижний регистр
df['Name'] = df['Name'].str.lower()
df['Nationality'] = df['Nationality'].str.lower()
df['Club'] = df['Club'].str.lower()
df['Preferred Foot'] = df['Preferred Foot'].str.lower()
df['Position'] = df['Position'].str.lower()
```

В названиях клубов имеются значения (1.), пробелы, пропуски, преходы.

```python
# Удаляем пробелы и переходы на новую строку.
df['Name'] = df['Name'].str.replace('\n', '')
df['Name'] = df['Name'].str.replace('\t', ' ')
df['Name'] = df['Name'].str.replace(' {2,}', ' ', regex=True)
df['Name'] = df['Name'].str.strip()

df['Club'] = df['Club'].str.replace('1.', '', regex=True)           # удаляем (1.) в названиях клубов
df['Club'] = df['Club'].str.replace('\n', '')
df['Club'] = df['Club'].str.replace('\t', ' ')
df['Club'] = df['Club'].str.replace(' {2,}', ' ', regex=True)
df['Club'] = df['Club'].str.strip()
```

### Задача №3. Напишите функцию, которая добавит колонку с разбиением возраста по группам: до 20, от 20 до 30, от 30 до 36 и старше 36. Посчитайте количество футболистов в каждой категории.

По описательной статистике возраст футболистов 16-44 лет.  
Возрастные группы:  
teenage - до 20 лет  
young_adult -от 20 до 30  
adult -от 30 до 36  
older_adult -старше 36

```python
df.loc[df['Age']<=20, 'Age_group'] = 'teenage'
df.loc[df['Age'].between(20,30), 'Age_group'] = 'young_adult'
df.loc[df['Age'].between(30,36), 'Age_group'] = 'adult'
df.loc[df['Age']>36, 'Age_group'] = 'older_adult'
```
